```python
#scrapping the data 
data = {"title":[], "review":[], "year":[], "price":[], "ratings":[]}
for a in range(1,76):
  pages = requests.get(f'https://www.amazon.com/s?k=amazon+shopping+online&i=stripbooks&rh=n%3A283155&dc&adgrpid=84486094307&gclid=CjwKCAjwwab7BRBAEiwAapqpTDYtTN8aym2q_rlYn1VdmofKw0_bzkd6xjaCl8cMY4syjbPZMKNMjxoCLIYQAvD_BwE&hvadid=393524136919&hvdev=c&hvlocphy=1010294&hvnetw=g&hvqmt=b&hvrand=2170543926145783317&hvtargid=kwd-469356929&hydadcr=22365_10729094&qid=1648813003&rnid=2941120011&ref=sr_pg_1',
                        headers=headers)
  
  soup = BeautifulSoup(pages.content, "html.parser")

  title = soup.find_all('h2', {"class": "a-size-mini a-spacing-none a-color-base s-line-clamp-2"})
  for i in title:
    for j in i.find_all('span', {'class':"a-size-medium a-color-base a-text-normal"}):
      data["title"].append(j.text)
  
      
 


  review=soup.find_all('div', {"class": "a-row a-size-small"})
  for i in review:
    for j in i.find_all('span', {"class": "a-size-base"}):
      data["review"].append(j.text)
    
  year = soup.find_all('div', {'class': "a-row" })
  for i in year:
    for j in i.find_all('span', {'class': 'a-size-base a-color-secondary a-text-normal'}):
      data['year'].append(j.text)

  price = soup.find_all('div', {'class':"a-row a-size-base a-color-base"})
  for i in price:
    for j in i.find_all('span', {'class': 'a-price-whole'}):
      data['price'].append(j.text)

  ratings = soup.find_all('div', {'class':"a-row a-size-small"})
  for i in ratings:
    for j in i.find_all('span', {'class': 'a-declarative'}):
      data['ratings'].append(j.text)
```

```python
#storing the data in a dataframe
books= pd.DataFrame.from_dict(data, orient='index')   
books.T
```
